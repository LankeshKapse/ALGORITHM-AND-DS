#### Insertion sort

```python
input_array = [1,4,5,3,7,9,8,6,0]
```

```python
length = len(input_array)
```

```python
for i in range(1,length):
    data = input_array[i]
    j = i - 1
    while j>=0 and input_array[j] >  data:
        #swapping
        input_array[j], input_array[j+1] = input_array[j+1], input_array[j]
        j = j - 1
        
```
